```python
import csv 
budget_file = ('budget_data.csv') 
outfile = open('budgetoutput.txt', 'w')

total_months = 0
net_total = 0
profit_loss_change = []
profitlosscolumn = []


with open(budget_file) as csvfile:
    csvreader = csv.reader(csvfile, delimiter =',')
    
    
    csvheader = next(csvreader)
        
    for row in csvreader:
        total_months += 1
        net_total += int(row[1])
        profitlosscolumn.append(float(row[1]))
    
    for i, (cur, nxt) in enumerate(zip(profitlosscolumn, profitlosscolumn[1:])):
        changes = nxt - cur
        profit_loss_change.append(changes)
    
    print(profit_loss_change)

        

  

       
        

        
    
    
        

## In order to calculate the average change, you need to do some iterations where you subtract the current profit/loss with the previous value and then take the average          


        
```
